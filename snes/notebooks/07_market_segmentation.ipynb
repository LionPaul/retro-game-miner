```python
import pandas as pd
import os

# ==============================================================================
# PIPELINE CONFIGURATION
# ==============================================================================
CONSOLE_NAME = "snes"

DATASETS_DIR = "datasets"
# ADJUSTMENT: Correct chained sequence tracking for the final Step 07
INPUT_FILENAME = f"6_{CONSOLE_NAME}_games_perfect_list.xlsx"   # Reading from Code 6
OUTPUT_FILENAME = f"7_{CONSOLE_NAME}_games_bestsellers.xlsx"   # Generating final base number 7

INPUT_PATH = os.path.join(DATASETS_DIR, INPUT_FILENAME)
OUTPUT_PATH = os.path.join(DATASETS_DIR, OUTPUT_FILENAME)
# ==============================================================================

def pipeline_market_segmentation(input_path, output_path):
    """
    Filters out games with zero recorded sales to isolate commercial hits.
    Acts as the specialized Gold Best-Sellers Layer of the Medallion Architecture.
    """
    print(f"💰 [CODE 07] Starting Market Data Isolation (Best-Sellers): {CONSOLE_NAME.upper()}")

    # 1. Pipeline gatekeeper check
    if not os.path.exists(input_path):
        print(f"❌ Critical Error: Input file '{input_path}' not found. Please run Notebook 06 first.")
        return None

    # 2. Load the definitive integrated dataset (Gold General Layer)
    df_perfect = pd.read_excel(input_path)
    total_analyzed_games = len(df_perfect)
    print(f"📊 Perfect dataset loaded successfully: {total_analyzed_games} games with scores and categories.")

    # 3. Market Filter: Retain only records that broke the 1 million copies threshold
    print("🎯 Filtering titles with active commercial metrics (Sales > 0)...")
    df_bestsellers = df_perfect[df_perfect['Copies_Sold_Millions'] > 0].copy()

    # Sorts the DataFrame by sales volume so market leaders sit right at the top
    df_bestsellers = df_bestsellers.sort_values(by='Copies_Sold_Millions', ascending=False)

    total_bestsellers = len(df_bestsellers)

    # 4. Persistence of the specialized Business Artifact
    df_bestsellers.to_excel(output_path, index=False)

    print("\n" + "═"*50)
    print(f"📊 MARKET ANALYSIS PIPELINE CLOSED SUCCESSFULLY!")
    print(f"🏆 Out of {total_analyzed_games} evaluated titles, {total_bestsellers} are official global Best-Sellers.")
    print(f"💾 Elite commercial dataset saved at: {output_path}")
    print("═"*50 + "\n")

    # Returns immediate visual feedback of market leaders to the Colab inspector
    print(f"🥇 TOP 5 COMMERCIAL HIT RECORDS FOR {CONSOLE_NAME.upper()}:")
    return df_bestsellers[['T1_Clean', 'Copies_Sold_Millions', 'IGDB_Score', 'Category_Primary']].head(5)

# Trigger the ultimate segmentation engine
df_bestsellers_preview = pipeline_market_segmentation(INPUT_PATH, OUTPUT_PATH)
```
